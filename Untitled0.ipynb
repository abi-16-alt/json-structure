```python
import fitz
import spacy
import json
import re
from datetime import datetime

import nltk
from nltk.corpus import wordnet
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tag import pos_tag
from nltk.stem import WordNetLemmatizer

from gensim.models import Word2Vec
from sentence_transformers import SentenceTransformer, util

nlp        = spacy.load("en_core_web_sm")
lemmatizer = WordNetLemmatizer()

print("✅ All imports successful.")

```

```python
ARTICLE_MARKERS = [
    "The Semantic Web", "Rip Van Twinkle", "Behind Enemy Lines",
    "The Arctic Oil", "Warp Drive Underwater",
    "What's Wrong with This Picture", "Project Skyhook",
    "Mad Cow", "Taken to Heart", "A Host with Infectious Ideas",
    "The Erotic-Fierce People", "Overview", "Letters",
    "50 and 100 Years Ago", "News Scan", "Explorations",
    "Technology and Business", "Working Knowledge", "Innovations",
]

def split_into_article_flows(text, markers):
    sections      = {}
    current_name  = "Front Matter"
    current_lines = []
    for line in text.split('\n'):
        stripped = line.strip()
        matched  = False
        for marker in markers:
            if marker.lower() in stripped.lower() and len(stripped) < 120:
                if current_lines:
                    sections[current_name] = ' '.join(current_lines)
                current_name  = marker
                current_lines = []
                matched = True
                break
        if not matched and stripped:
            current_lines.append(stripped)
    if current_lines:
        sections[current_name] = ' '.join(current_lines)
    return sections

article_flows = split_into_article_flows(full_text, ARTICLE_MARKERS)
print(f"\n📚 Article flows: {len(article_flows)}")
for name, content in article_flows.items():
    print(f"   ▸ '{name}' — {len(content):,} chars")

```

```python
def tokenise_flow(sections):
    flow_sentences = {}
    for name, content in sections.items():
        content   = re.sub(r'\s+', ' ', content).strip()
        sentences = [s.strip() for s in sent_tokenize(content) if len(s.strip()) > 20]
        flow_sentences[name] = sentences
    return flow_sentences

flow_sentences = tokenise_flow(article_flows)
print("\n📝 Sentences per flow:")
for name, sents in flow_sentences.items():
    print(f"   ▸ '{name}': {len(sents)} sentences")


```

```python
print("\n⏳ Loading sentence embedding model...")
embed_model = SentenceTransformer('all-MiniLM-L6-v2')
print("✅ Loaded.")

def order_sentences(sentences):
    if len(sentences) <= 1:
        return sentences, list(range(len(sentences)))
    embeddings = embed_model.encode(sentences, convert_to_tensor=True)
    n = len(sentences)
    visited = [False] * n
    order   = [0]
    visited[0] = True
    for _ in range(n - 1):
        curr   = order[-1]
        best_j = max((j for j in range(n) if not visited[j]),
                     key=lambda j: util.cos_sim(embeddings[curr], embeddings[j]).item())
        order.append(best_j)
        visited[best_j] = True
    return [sentences[i] for i in order], order

print("\n🔀 Ordering sentences...")
ordered_flows = {}
for name, sents in flow_sentences.items():
    if sents:
        ordered, idx = order_sentences(sents)
        ordered_flows[name] = {"original": sents, "ordered": ordered, "order_idx": idx}
        print(f"   ✅ '{name}': {len(ordered)} sentences")
```

```python
def get_wn_pos(tag):
    if tag.startswith('J'): return wordnet.ADJ
    if tag.startswith('V'): return wordnet.VERB
    if tag.startswith('R'): return wordnet.ADV
    return wordnet.NOUN

def enrich_with_wordnet(word, tag):
    wn_pos  = get_wn_pos(tag)
    lemma   = lemmatizer.lemmatize(word.lower(), wn_pos)
    synsets = wordnet.synsets(lemma, pos=wn_pos)
    if not synsets:
        return None
    primary  = synsets[0]
    synonyms = list({l.name().replace('_',' ') for s in synsets[:3]
                     for l in s.lemmas() if l.name().lower() != lemma})[:5]
    hypernyms = [h.lemma_names()[0].replace('_',' ') for h in primary.hypernyms()][:3]
    return {"lemma": lemma, "definition": primary.definition(),
            "synonyms": synonyms, "hypernyms": hypernyms}

print("✅ WordNet helpers ready.")
```

```python
# Fix: download the correct tagger name for newer NLTK versions
import nltk
nltk.download('averaged_perceptron_tagger_eng')

# Then re-run Cell 8 (extract_entities) as normal
ENTITY_TYPES = ['PERSON','ORG','GPE','PRODUCT','WORK_OF_ART','EVENT','NORP','FAC']

def make_business_rules(name):
    pfx = name[:4].upper()
    return [
        {"ruleId": f"BR-{pfx}-001",
         "description": f"{name} must have a unique identifier",
         "enforcement": "primaryKeyConstraint"},
        {"ruleId": f"BR-{pfx}-002",
         "description": f"{name} must be linked to a valid source article",
         "enforcement": "foreignKeyConstraint"},
        {"ruleId": f"BR-{pfx}-003",
         "description": f"{name} name must not be empty",
         "expression":  f"length({name}.name) > 0"},
        {"ruleId": f"BR-{pfx}-004",
         "description": f"All required attributes of {name} must be present",
         "enforcement": "notNullConstraint"},
    ]

def extract_entities(sentences, article_name):
    entities = {}
    for sent_idx, sent in enumerate(sentences):
        doc_spacy = nlp(sent)
        tokens    = word_tokenize(sent)
        pos_tags  = pos_tag(tokens)

        for ent in doc_spacy.ents:
            if ent.label_ not in ENTITY_TYPES:
                continue
            name = ent.text.strip()
            if name in entities:
                continue

            # Attributes (nouns → WordNet enriched)
            attrs = {}
            for tok, tag in pos_tags:
                if tag.startswith('N') and len(tok) > 3:
                    info = enrich_with_wordnet(tok, tag)
                    if info:
                        attrs[tok] = {
                            "type": "string",
                            "description": info["definition"],
                            "synonyms":    info["synonyms"],
                            "hypernyms":   info["hypernyms"],
                        }
                    if len(attrs) >= 6:
                        break

            # Actions (verbs)
            actions = []
            for token in doc_spacy:
                if token.pos_ != "VERB" or len(token.lemma_) <= 2:
                    continue
                is_inter = any(t.ent_type_ in ENTITY_TYPES for t in token.children)
                targets  = [t.text for t in token.children if t.ent_type_ in ENTITY_TYPES]
                act_obj  = {
                    "actionName":  token.lemma_,
                    "type":        "interaction" if is_inter else "local",
                    "description": f"{token.lemma_} performed by {name}",
                }
                if targets:
                    act_obj["targetEntity"] = targets[0]
                if not any(a["actionName"] == act_obj["actionName"] for a in actions):
                    actions.append(act_obj)
                if len(actions) >= 5:
                    break

            entities[name] = {
                "entity": {
                    "name":        name,
                    "type":        ent.label_,
                    "description": f"{ent.label_} entity from '{article_name}'",
                    "version":     "1.0",
                    "sourceFlow":  article_name,
                },
                "attributes":    attrs,
                "actions":       actions,
                "relationships": [],
                "businessRules": make_business_rules(name),
                "audit": {
                    "createdBy":        "sciam_nlp_pipeline",
                    "createdDate":      datetime.now().isoformat(),
                    "lastModifiedBy":   "system",
                    "lastModifiedDate": datetime.now().isoformat(),
                },
                "_meta": {"sentenceIndex": sent_idx, "sourceSentence": sent[:200]},
            }
    return entities

print("\n🔍 Extracting entities (Doc-72 schema)...")
flow_entities = {}
for article_name, data in ordered_flows.items():
    ents = extract_entities(data["ordered"], article_name)
    flow_entities[article_name] = ents
    print(f"   ✅ '{article_name}': {len(ents)} entities")


```

```python


def detect_relationships(sentences, entities):
    names = list(entities.keys())
    interactions = []
    for idx, sent in enumerate(sentences):
        present = [e for e in names if e.lower() in sent.lower()]
        for i in range(len(present)):
            for j in range(i+1, len(present)):
                interactions.append({
                    "sourceEntity":    present[i],
                    "targetEntity":    present[j],
                    "relationshipType":"co-occurrence",
                    "temporalOrder":   idx,
                    "context":         sent[:200],
                })
                rel = {"entity": present[j], "relationshipType": "co-occurrence",
                       "temporalOrder": idx,
                       "description": f"Co-occurs with {present[j]}"}
                if rel not in entities[present[i]]["relationships"]:
                    entities[present[i]]["relationships"].append(rel)
    return interactions

print("\n🔗 Detecting relationships...")
flow_interactions = {}
for article_name, data in ordered_flows.items():
    if article_name in flow_entities:
        ints = detect_relationships(data["ordered"], flow_entities[article_name])
        flow_interactions[article_name] = ints
        if ints:
            print(f"   ✅ '{article_name}': {len(ints)} interactions")
```

```python
def assemble_json(article_name, ordered_sents, entities, interactions):
    top_ents = sorted(entities.items(),
                      key=lambda x: len(x[1]["attributes"]), reverse=True)[:12]
    return {
        "flowName":         article_name,
        "totalSentences":   len(ordered_sents),
        "summarySentences": ordered_sents[:3],
        "entities":         [e for _, e in top_ents],
        "interactions":     interactions[:10],
        "generatedAt":      datetime.now().isoformat(),
        "schema":           "Doc-72 v1.0",
    }

print("\n📦 Assembling final JSON...")
final_output = {}
for article_name in ordered_flows:
    if article_name in flow_entities:
        final_output[article_name] = assemble_json(
            article_name,
            ordered_flows[article_name]["ordered"],
            flow_entities[article_name],
            flow_interactions.get(article_name, [])
        )

print(f"✅ JSON ready for {len(final_output)} flows.")
sample = list(final_output.keys())[0]
print(f"\n📋 Preview — '{sample}':")
print(json.dumps(final_output[sample], indent=2)[:2000])
```

```python
def build_corpus(ordered_flows, flow_entities):
    corpus = []
    for article_name, data in ordered_flows.items():
        # Sentence-level tokens
        for sent in data["ordered"]:
            toks = [w.lower() for w in word_tokenize(sent)
                    if w.isalpha() and len(w) > 2]
            if toks:
                corpus.append(toks)
        # Structured entity context
        if article_name in flow_entities:
            for ent_data in flow_entities[article_name].values():
                ctx = ([ent_data["entity"]["name"].lower().replace(' ','_')] +
                       list(ent_data["attributes"].keys()) +
                       [a["actionName"] for a in ent_data["actions"]])
                if ctx:
                    corpus.append(ctx)
    return corpus

training_corpus = build_corpus(ordered_flows, flow_entities)
print(f"\n📚 Training corpus: {len(training_corpus)} sequences")

print("⏳ Training WordScene (Word2Vec)...")
wordscene = Word2Vec(
    sentences   = training_corpus,
    vector_size = 100,
    window      = 5,
    min_count   = 2,
    workers     = 4,
    epochs      = 30,
    sg          = 1,     # skip-gram
)
print(f"✅ WordScene trained!  Vocab: {len(wordscene.wv)} words")

# Sample similarity checks
test_words = ['science','research','technology','data','system']
print("\n🔍 Similarity examples:")
for word in test_words:
    if word in wordscene.wv:
        similar = wordscene.wv.most_similar(word, topn=5)
        print(f"\n  '{word}':")
        for w, sc in similar:
            print(f"     {w:<22} {sc:.4f}")

wordscene.save("wordscene_sciam.model")
print("\n💾 Saved: wordscene_sciam.model")

```

```python
def query_word(word):
    print(f"\n{'='*55}\n  QUERY: '{word}'\n{'='*55}")
    # WordNet
    synsets = wordnet.synsets(word)
    if synsets:
        syns = [l.name() for s in synsets[:2] for l in s.lemmas()][:6]
        hyps = [h.lemma_names()[0] for h in synsets[0].hypernyms()]
        print(f"\n📖 WordNet:\n   Def  : {synsets[0].definition()}")
        print(f"   Syns : {', '.join(syns)}")
        if hyps: print(f"   Hyps : {', '.join(hyps)}")
    else:
        print("   (no WordNet entry)")
    # WordScene
    if word in wordscene.wv:
        print(f"\n🗺️  WordScene:")
        for w, sc in wordscene.wv.most_similar(word, topn=5):
            print(f"   {w:<22} {sc:.4f}")
    else:
        print(f"\n   ('{word}' not in WordScene vocab)")

query_word("web")
query_word("computer")
query_word("brain")
```
